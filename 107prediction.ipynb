```python
dict_1=df1.to_dict(orient='list')
```

```python
case_id = dict_1['CASE_ID']
```

```python
workflow = dict_1['WRKFLW_TSK_MAP_DETAILED']
```

```python
print(type(workflow))
print(len(workflow))
```

```python
from collections import defaultdict

case_wf_dict = defaultdict(list)

for i in range(len(case_id)):
    item = case_id[i]
    case_wf_dict[item].append(workflow[i])
    
print(len(case_wf_dict))
```

```python
df2 = df2.set_index('CASE_ID')
```

```python
df2.head(5)
```

```python
df2_col_name = df2.columns.values.tolist()
print(df2_col_name)
```

```python
for key in case_wf_dict.keys():
    for process in case_wf_dict[key]:
        if process in df2_col_name:
            df2.loc[key, process] = 1
        else:
            continue
```
